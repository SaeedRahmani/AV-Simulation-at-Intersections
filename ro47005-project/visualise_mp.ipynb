```python
import matplotlib.pyplot as plt
import numpy as np
import pickle

% matplotlib widget
```

```python
def open_pkl(filename):
    with open(filename, 'rb') as file:
        return pickle.load(file)


plt.close
plt.figure()
plt.axis('equal')
# plt.ylim(0, 1)
filenames = [
    'data/motion_primitives/straight.pkl',
    'data/motion_primitives/left1.pkl',
    'data/motion_primitives/left2.pkl'
]

for filename in filenames:
    points = open_pkl(filename).points
    points = np.array(points)
    plt.plot(points[:, 1], points[:, 0], color='b')
    # plt.plot(-points[:, 1], points[:, 0])

plt.show()
```
