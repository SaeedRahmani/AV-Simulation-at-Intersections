```python
import matplotlib.pyplot as plt
import numpy as np
import pickle

%matplotlib widget
```

```python
def open_pkl(filename):
    with open(filename, 'rb') as file:
        return pickle.load(file)


plt.close
plt.figure()
plt.axis('equal')
# plt.ylim(0, 1)
filenames = [
    'data/motion_primitives_prius/straight.pkl',
    'data/motion_primitives_prius/left1.pkl',
    'data/motion_primitives_prius/left2.pkl'
]

for filename in filenames:
    points = open_pkl(filename).points
    points = np.array(points)
    if filename == filenames[0]:
        plt.scatter(points[:, 0], points[:, 1], color='b', label='MP from pybullet')
    else:
        plt.scatter(points[:, 0], points[:, 1], color='b')
    # plt.plot(-points[:, 1], points[:, 0])

filenames = [
    'data/motion_primitives_bicycle_model/straight.pkl',
    'data/motion_primitives_bicycle_model/left1.pkl',
    'data/motion_primitives_bicycle_model/left2.pkl'
]

for filename in filenames:
    points = open_pkl(filename).points
    points = np.array(points)
    if filename == filenames[0]:
        plt.scatter(points[:, 0], points[:, 1], color='r', label='MP from our Bicycle model')
    else:
        plt.scatter(points[:, 0], points[:, 1], color='r')
    # plt.plot(-points[:, 1], points[:, 0])

filenames = [
    'data/motion_primitives_bicycle_model_real_size/straight.pkl',
    'data/motion_primitives_bicycle_model_real_size/left1.pkl',
    'data/motion_primitives_bicycle_model_real_size/left2.pkl',
    'data/motion_primitives_bicycle_model_real_size/left3.pkl'
]

for filename in filenames:
    points = open_pkl(filename).points
    points = np.array(points)
    if filename == filenames[0]:
        plt.scatter(points[:, 0], points[:, 1], color='g', label='MP from our Bicycle model')
    else:
        plt.scatter(points[:, 0], points[:, 1], color='g')
    # plt.plot(-points[:, 1], points[:, 0])

plt.legend()
plt.show()
```
